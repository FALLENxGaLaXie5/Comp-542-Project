```python
import pandas as pd
import plotly.express as px
df = pd.read_csv("FireDataCleaningProgress.csv");
print(df.head(3));
```

```python
## Drop invaild(0) latitude/longitude data here
DroppedInvaildLatLong = df.loc[(df['Latitude'] != 0) & (df['Longitude'] != 0)]
print(DroppedInvaildLatLong)
```

```python
## Box California Geographic shape
## left to right, change in Longitude
## up and down, change in Latitude
##                 Latitude Longitude
## Leftest point 34.231887,-114.138479
## rightest point  40.286597,-124.750379
## Upperest Point 41.973466,-124.588296
## Lowest Point  32.541459,-117.106648
DroppedInvaildPoints = df.loc[((df['Latitude'] > 32.541459) & (df['Latitude'] < 41.973466)) &((df['Longitude'] > -124.750379) & (df['Longitude'] < -114.138479))]
Rearragneddf = DroppedInvaildPoints.sort_values(['Latitude'], ascending =True)
print("\nDisplay upper Latitude limits:")
print(Rearragneddf.head(1))
print("\nDisplay lower Latitude limits:")
print(Rearragneddf.tail(1))
print("\nDisplay upper Longitude limits:")
FinalArrangement = DroppedInvaildPoints.sort_values(['Longitude'], ascending =True)
print(FinalArrangement.head(1))
print("\nDisplay lower Longitude limits:")
print(FinalArrangement.tail(1))
print("\nDisplay FinalArrangement dataframe:")
print(FinalArrangement)
```

```python
##mapping all California fire incidents on the map
fig = px.scatter_geo(FinalArrangement,lat='Latitude',lon='Longitude')
fig.update_layout(title = 'Fire Data MappingV.1', title_x=0.5)
fig.show()
```

```python
##Remove outLiner then remap
new_df=FinalArrangement.drop(FinalArrangement[(FinalArrangement['Latitude'] == 38.838) & (FinalArrangement['Longitude'] == -116.61)].index)
fig = px.scatter_geo(new_df,lat='Latitude',lon='Longitude')
fig.update_layout(title = 'Fire Data Mapping V.2', title_x=0.5)
fig.show()
print(new_df)
```
