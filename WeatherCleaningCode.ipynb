```python
import pandas as pd
import plotly.express as py
dataFrame = pd.read_csv("Combined Weather Data 2013-2019.csv")
print(dataFrame.head(3))
print(dataFrame.tail(3))
```

```python
#dataFrame.insert(19, "Station_Num", 1)
dataFrame = dataFrame.assign(Station_Num=1)
```

```python
print(dataFrame)
```

```python
import numpy as np
dataFrame['Station_Num'] = np.where(((dataFrame['LONGITUDE'] == 33.44) & (dataFrame['LATITUDE '] == -117.19)), 2, dataFrame['Station_Num'])
print(dataFrame)
```

```python
dataFrame['Station_Num'] = np.where(((dataFrame['LONGITUDE'] == -119.82) & (dataFrame['LATITUDE '] == 37.76)), 7, dataFrame['Station_Num'])
print(dataFrame.tail(3))
```

```python
##Append station numbers to the rest of the stations
dataFrame['Station_Num'] = np.where(((dataFrame['LONGITUDE'] == -117.19) & (dataFrame['LATITUDE '] == 33.44)), 2, dataFrame['Station_Num'])
dataFrame['Station_Num'] = np.where(((dataFrame['LONGITUDE'] == -120.88) & (dataFrame['LATITUDE '] == 37.24)), 3, dataFrame['Station_Num'])
dataFrame['Station_Num'] = np.where(((dataFrame['LONGITUDE'] == -122.61) & (dataFrame['LATITUDE '] == 40.65)), 4, dataFrame['Station_Num'])
dataFrame['Station_Num'] = np.where(((dataFrame['LONGITUDE'] == -119.88) & (dataFrame['LATITUDE '] == 34.41)), 5, dataFrame['Station_Num'])
dataFrame['Station_Num'] = np.where(((dataFrame['LONGITUDE'] == -117.14) & (dataFrame['LATITUDE '] == 36.6)), 6, dataFrame['Station_Num'])
```

```python
fireDF = pd.read_csv("masterFire.csv")
print(fireDF.head(3))
```

```python
#dataFrame.insert(20, "Fire", 0)
dataFrame = dataFrame.assign(Fire=0)
```

```python
print(dataFrame)
```

```python
#Add fires into weather data as a '1'
i = 0
for indexWeather, rowWeather in dataFrame.iterrows():
    for indexFire, rowFire in fireDF.iterrows():
        if (rowWeather['Station_Num'] == rowFire['LocationIndex']) & (rowWeather['LST_DATE '] == rowFire['Started']):
            dataFrame.at[i, 'Fire'] = 1
    i += 1
```

```python
#Delete any invalid rows with missing data
#possibly just replace data so we have more to go off of
droppedInvalidWeather = dataFrame.loc[(dataFrame['T_DAILY_MAX '] > -10) & (dataFrame[' T_DAILY_AVG'] > -10) & (dataFrame[' P_DAILY_CALC '] > -10) & (dataFrame['SOLARAD_DAILY '] > -10) & (dataFrame['SUR_TEMP_DAILY_MAX '] > -10) & (dataFrame['SUR_TEMP_DAILY_AVG '] > -10) & (dataFrame['RH_DAILY_MIN '] > -10) & (dataFrame['RH_DAILY_AVG '] > -10) & (dataFrame['SOIL_MOISTURE_5_DAILY '] > -10) & (dataFrame['SOIL_MOISTURE_10_DAILY '] > -10) & (dataFrame['SOIL_MOISTURE_20_DAILY '] > -10) & (dataFrame['SOIL_MOISTURE_50_DAILY '] > -10) & (dataFrame['SOIL_MOISTURE_100_DAILY '] > -10) & (dataFrame['SOIL_TEMP_5_DAILY '] > -10) & (dataFrame['SOIL_TEMP_10_DAILY '] > -10)]
print(droppedInvalidWeather)
```

```python
droppedInvalidWeather.to_csv('cleaned_data1.csv')
```
